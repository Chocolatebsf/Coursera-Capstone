```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

Download the data from the wiki

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
d=pd.read_html(url)
df=d[0]
df
```

Remove rows of Not Assigned in the column Borough and Neighborhood

```python
df1=df[df.Borough !='Not assigned']
df1.head()
```

```python
df3=df1.replace(to_replace="Not assigned",value="Queen's Park")
```

Group Neighbourhood by postcode and Borough

```python
df2=df3.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df2
```

```python
df2.shape
```

Download the CSV of the longtitude and latitude data

```python
url1='http://cocl.us/Geospatial_data'
df_longlat=pd.read_csv(url1)
```

```python
df_longlat
```

Join the longtitude and latitude dataframe with Postcode of Toronto

```python
df_combo=pd.concat([df2,df_longlat],axis=1)
```

Drop the repeater on post code

```python
df_combo=df_combo.drop('Postal Code',axis=1)
```

```python
df_combo
```

```python
print("City of Toronto has {} boroughs and {} neighbourhoods.".format(len(df_combo['Borough'].unique()),df_combo.shape[0]))
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_combo['Latitude'], df_combo['Longitude'], df_combo['Borough'], df_combo['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
